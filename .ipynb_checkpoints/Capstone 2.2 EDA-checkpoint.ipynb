now my mind wonders to outliers. there are quite a bit of them, but i dont think they concern me at this point.

```python
# Dictionary to hold the count of outliers for each column
outlier_counts = {}

# Calculating IQR and determining outliers for each numeric column
for col in numeric_data:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((data[col] < (Q1 - 1.5 * IQR)) | (data[col] > (Q3 + 1.5 * IQR)))
    outlier_counts[col] = outliers.sum()

# Plotting the counts of outliers
plt.figure(figsize=(10, 6))
sns.barplot(x=list(outlier_counts.keys()), y=list(outlier_counts.values()))
plt.title('Count of Outliers per Column')
plt.ylabel('Number of Outliers')
plt.xlabel('Columns')
plt.xticks(rotation=45)
plt.show()
```

but now i am curious about how often unknown matches with unknown in other categories.

```python
is_unknown = data.applymap(lambda x: x == 'unknown')

# Initialize an empty DataFrame to store co-occurrence counts
co_occurrence_counts = pd.DataFrame(index=data.columns, columns=data.columns, data=0)

# Count co-occurrences
for col1 in data.columns:
    for col2 in data.columns:
        # Count rows where both columns have 'unknown'
        count = len(data[(is_unknown[col1]) & (is_unknown[col2])])
        co_occurrence_counts.loc[col1, col2] = count

plt.figure(figsize=(12, 10))
sns.heatmap(co_occurrence_counts, annot=True, cmap='coolwarm', fmt='d')
plt.title('Co-occurrence Counts of Unknown Values')
plt.show()
```

```python
# Increase the figure size
plt.figure(figsize=(15, 12))

# Mask to hide the upper triangle, including the diagonal
mask = np.triu(np.ones_like(co_occurrence_counts, dtype=bool))

# Create the heatmap with smaller annotation font size and a mask
sns.heatmap(co_occurrence_counts, annot=True, cmap='coolwarm', fmt='d', annot_kws={'size': 8}, mask=mask)

plt.title('Co-occurrence Counts of Unknown Values')
plt.show()
```

super wow. i am now feeling even more justified i my desire to drop girdle min and max: not only do they create an 'unknown' match with themselves, they also match 'unknown' entries in three other fields.

checking distributions, this part may need some work

```python
from scipy.stats import kstest, norm

results = []

# Iterate through each column
for col in data.columns:
    if pd.api.types.is_numeric_dtype(data[col]):
        # Calculate the necessary statistics
        mean, std = data[col].mean(), data[col].std()
        
        # Perform the K-S test against a normal distribution
        stat, p_value = kstest(data[col], 'norm', args=(mean, std))

        # Append the results
        results.append({'Column': col, 'Statistic': stat, 'P-value': p_value})

# Convert the results to a DataFrame
ks_test_results = pd.DataFrame(results)

ks_test_results
```

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram and KDE of {col}')
    
    plt.subplot(1, 2, 2)
    stats.probplot(data[col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {col} vs Normal Distribution')

    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import kstest, expon

results = []

# Iterate through each column
for col in data.columns:
    if pd.api.types.is_numeric_dtype(data[col]):
        # Calculate the scale parameter (lambda) as the inverse of the mean
        lambda_est = 1 / data[col].mean() if data[col].mean() > 0 else 0

        # Perform the K-S test (if mean is positive and column not empty)
        if lambda_est > 0 and not data[col].isnull().all():
            stat, p_value = kstest(data[col], 'expon', args=(0, lambda_est))
            results.append({'Column': col, 'K-S Statistic': stat, 'P-value': p_value})
        else:
            results.append({'Column': col, 'K-S Statistic': 'N/A', 'P-value': 'N/A'})

# Convert the results to a DataFrame for easier viewing
ks_test_results = pd.DataFrame(results)
ks_test_results
```

```python
# Select only numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    plt.figure(figsize=(12, 6))
    
    # Plot histogram and KDE
    plt.subplot(1, 2, 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram and KDE of {col}')
    
    # Generate a Q-Q plot against the exponential distribution
    plt.subplot(1, 2, 2)
    stats.probplot(data[col], dist="expon", plot=plt)
    plt.title(f'Q-Q Plot of {col} vs Exponential Distribution')

    plt.tight_layout()
    plt.show()
```

now looking at some chi squared tests on the categorical data

```python
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

from scipy.stats import chi2_contingency
import pandas as pd

# Store results
results = []

# Iterate over all pairs of categorical columns
for i in range(len(categorical_cols)):
    for j in range(i + 1, len(categorical_cols)):
        # Create a contingency table
        crosstab = pd.crosstab(data[categorical_cols[i]], data[categorical_cols[j]])
        
        # Perform the Chi-squared test
        chi2, p, dof, expected = chi2_contingency(crosstab)
        
        # Append results
        results.append({
            'Variable 1': categorical_cols[i],
            'Variable 2': categorical_cols[j],
            'Chi2 Statistic': chi2,
            'P-value': p
        })

# Convert the results to a DataFrame
chi_squared_results = pd.DataFrame(results)
```

```python
pivot_table = chi_squared_results.pivot('Variable 1', 'Variable 2', 'P-value')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='viridis', fmt=".2g")
plt.title('P-values of Chi-Squared Tests for Categorical Variable Pairs')
plt.show()
```
